```python
import numpy as np
l = np.linspace(60, 10000, 100)
```

```python
def k(l, z):
    
    return l/comving_angular(z)  
```

```python
import matplotlib.pyplot as plt
%pylab inline
```

```python
plt.figure(figsize=(20,20))
for z in [0.5, 0.8 , 1. , 1.5]:
    plt.loglog(l, k(l,z), lw = 3, label = "redshift = "+str(z))
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
plt.legend(loc = 'best', fontsize = 30)
plt.xlabel("ell", fontsize = 40)
plt.ylabel("k [1/Mpc]", fontsize = 40)
```
